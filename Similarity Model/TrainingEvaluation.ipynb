```python
from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv('trainingDataSet.csv', sep=",", encoding='unicode_escape')
X, y, z = df[['Sentence1','Sentence2']], df['Similarity Point'].astype('float32'), df['Similarity Type']

X_train, X_test, y_train, y_test, z_train, z_test = train_test_split(X, y, z, test_size=0.30, random_state=42)


# print('hahah', z_test[42882])
# for i in range(len(y)):
#   if i in y_test:
#     if z_test[i] == 3:
#       print(type(X_test['Sentence1'][i]))

# haha = y_train[860]
# haha2 = float(y_train[860])
# print(type(haha))
# print('__getitem__' in dir(haha))
# print('__getitem__' in dir(haha2))
# print(len(y))
# print(haha)
```

```python
 #train and test the Model with the whole data set

from sentence_transformers import evaluation, SentenceTransformer, losses, InputExample
from torch.utils.data import DataLoader

sentences1 = []
sentences2 = []
scores = []

model = SentenceTransformer('distilbert-base-nli-mean-tokens', device='cuda')

for i in range(len(y)):
  if i in y_test:
    sentences1.append(X_test['Sentence1'][i])

for i in range(len(y)):
  if i in y_test:
    sentences2.append(X_test['Sentence2'][i])

for i in range(len(y)):
  if i in y_test:
    scores.append(y_test[i])

evaluator = evaluation.EmbeddingSimilarityEvaluator(sentences1, sentences2, scores)

# ... Your other code to load training data

train_examples = []
for i in range(len(y)):
    if i in y_train:
      inp = InputExample(texts=[X_train['Sentence1'][i], X_train['Sentence2'][i]], label=float(y_train[i].item()))
      train_examples.append(inp)

train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)


train_loss = losses.CosineSimilarityLoss(model)

model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=30, warmup_steps=100, evaluator=evaluator, evaluation_steps=500, output_path='model30')
```
